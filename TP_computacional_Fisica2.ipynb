Con todas estas consideraciones tomamos la siguiente ecuaci√≥n para la masa A en coordenadas polares
$$Ec. 1$$
$$mL \ddot {\theta_a}=  -mg \theta_a + kL(\theta_b - \theta_a) +kL(\theta_c - \theta_a)$$  

Y hasta ac√° el resto del problema es an√°logo, de hecho se puede notar que lo que elegimos como "masa a" es arbitrario porque los 3 cuerpos se encuentran en situaciones id√©nticas. Al reordenar la ecuaci√≥n y si lo extendemos a las masas restantes, obtenemos este sistema de ecuaciones de Newton:

 $$Ec. 2$$   $$\ddot \theta_a=(-\frac {g} {L} -2 \frac {k} {m})\theta_a+ \frac {k} {m} (\theta_b +\theta_c) $$
$$\ddot \theta_b=(-\frac {g} {L} -2 \frac {k} {m})\theta_b+ \frac {k} {m} (\theta_a +\theta_c) $$
$$\ddot \theta_c=(-\frac {g} {L} -2 \frac {k} {m})\theta_c+ \frac {k} {m} (\theta_a +\theta_b) $$

Ecuaciones que se pueden llevar a forma matricial. Notamos que habr√° una diagonal principal y 2 adyacentes:

$$ \overrightarrow{\ddot \theta} = - \begin{bmatrix}
\frac{2k}{m} + \frac{g}{L} & -\frac{k}{m} & -\frac{k}{m} \\
-\frac{k}{m} & \frac{2k}{m} + \frac{g}{L} & -\frac{k}{m} \\
-\frac{k}{m} & -\frac{k}{m} & \frac{2k}{m} + \frac{g}{L}
\end{bmatrix} \overrightarrow{\theta} $$


Procedemos a calcular su determinante con $$Ec. 3$$ $$\lambda =\omega ^2$$

$$det\begin{pmatrix}
\frac{2k}{m} + \frac{g}{L} - \lambda& -\frac{k}{m} & -\frac{k}{m} \\
-\frac{k}{m} & \frac{2k}{m} + \frac{g}{L} - \lambda& -\frac{k}{m} \\
-\frac{k}{m} & -\frac{k}{m} & \frac{2k}{m} + \frac{g}{L} - \lambda
\end{pmatrix} = 0$$

El resultado del determinante nos di√≥ 2 autovalores (frecuencias naturales) y 3 autovectores lo cual, una vez m√°s , refleja el exceso de simetr√≠a del problema con sus "extremos" unidos entre si.
$$Ec. 4$$ $$ \lambda_1=\frac {g} {L}$$ con el autovector que relaciona las amplitudes $v_1$ = (1,1,1)

```python
#PARA EL LECTOR: ingrese el valor de n masas que prefiera entre 4 y 10
n = 5
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc # librer√≠a de animaciones
rc('animation', html='html5') #para poder ver la animacion en formato inline
from IPython.display import HTML

m = 1 #[kg]
k = 0.9 #[N/m^2]
R = 10 #[m]
L = 5 #[m]
l0 = 5 #(R*2*np.pi)/n #[m]
g = 9.8 #[g/s^2]
N = np.arange(1, n+1)
#N = ['a', 'b', 'c', 'd', 'e']
# thethaPrimaPrimaN =-((2k/m)+(g/L))thethaN + (k/m)(thethaN+1 + thethaN-1)

dp = (2*k/m) + (g/L)
do = (k/m)

print("Las ecuaciones de Newton para n = " + str(n) + " part√≠culas son:")

i=0
while i < len(N):
  if i == n-1:
    print("Thetha_" + str(N[i]) + "'' = " + str(dp) + " * Thetha_" + str(N[i]) + " + " + str(do) + " * ( Theta_" + str(N[0]) + " + Thetha_" + str(N[i-1]) + ")")
    break

  print("Thetha_" + str(N[i]) + "'' = " + str(dp) + " * Thetha_" + str(N[i]) + " + " + str(do) + " * ( Theta_" + str(N[i+1]) + " + Thetha_" + str(N[i-1]) + ")")
  i += 1
```

**2. Resuelva el sistema de autovectores y autovalores para hallar las frecuencias y los modos
normales de oscilaci√≥n.**

```python
def generaM(n, dp, do):
  M = np.empty((n, n))

  i=0
  while i < len(N):
    a = np.zeros((n,), dtype = float)

    if i == n-1:
      a[i] = dp;   a[0] = -do; a[i-1] = -do
      #print(a)
      M[i] = a
      break

    a[i] = dp;   a[i+1] = -do; a[i-1] = -do
    #print(a)
    M[i] = a

    i += 1

  return M

print('Con estos valores para ' + str(n) + ' part√≠culas. La matriz M nos queda: ')
matriz = generaM(n, dp, do)
print(matriz)
print('Lo cual nos permite calcular sus frecuencias (ra√≠z cuadrada de los autovalores): ')
autoval, autovect = np.linalg.eigh(matriz)
autoval = np.sqrt(autoval)
omega=np.around(autoval,2)
vector_amplitud=np.around(autovect,2)
print(omega)
print('Y sus respectivos autovectores (modos normales): ')
print(vector_amplitud)
```

**3. Especifique el problema para n = 3 y verifique que recupera los resultados obtenidos en la
Parte 1**

```python
matrizN3 = generaM(3, dp, do)

print('Para el caso presentado de 3 part√≠culas. La matriz M nos queda: ')
print(matrizN3)
print('Cuyos autovalores (frecuencias) son: ')
autoval3, autovect3 = np.linalg.eigh(matrizN3)
autoval3 = np.sqrt(autoval3)
omegaN3=np.around(autoval3,2)
vector_amplitudN3=np.around(autovect3,2)
print(omegaN3)
print('Y sus autovectores (modos normales) son: ')
print(vector_amplitudN3)
```

Donde, como anticipamos, hay 2 autovalores iguales y en el caso de los autovectores hay una combinacion lineal de los que obtuvimos.

#5. masa 10 veces mas grande

Al agregar una masa considerablemente mayor al resto , respecto a las ecuaciones , no cambia mucho. Si le damos a la masa a un valor de 10m su ecuacion es la unica que cambia respecto a la **Ec. 7**.
$$Ec. 8$$ $$\ddot \theta a=(-\frac{g}{l}-\frac{k}{5m})\theta a+\frac{k}{10m}\theta b+\frac{k}{10m}\theta e$$
Y aunque esta sea la unica ecuacion del sistema de 5x5 que cambia , basta con eso para romper la simetria. Si calculamos los autovalores y autovectores de este nuevo sistema lo que se vera es que ya no hay autovalores repetidos sino que todos son distintos lo cual refleja nuestra simetria rota.

Algo para destacar de esto es que , aunque nuestra simetria se haya roto ,en el modo fundamental todas las masas se mueven en fase ¬øesto tiene sentido? ¬øno deberia la masa con mayor masa moverse menos? La respuesta nos la dan las mismas cuentas y es que no. En este modo no esperamos que los resortes que acoplan el sistema actuen porque no presentan estiramiento lo cual nos deja unicamente la ecuacion del pendulo cuya aceleracion no depende de la masa. Entonces tiene sentido que las masa $a$, $b$, $c$, $d$ y $e$ se muevan en fase.
